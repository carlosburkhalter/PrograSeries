```python
import numpy as np

data = np.genfromtxt('density_air.dat')
x1 = data[:,0]
y1 = data[:,1]

import matplotlib.pyplot as plt 

coeff = np.polyfit(x1,y1,3)
p = np.poly1d(coeff)
print(p) # prints the polynomial expression
y_fitted = p(x1)

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(x1, y1, 'r')
ax.plot(x1, y_fitted, 'b',dashes=[1, 8])
#ax.legend("data")
################################################

data2= np.genfromtxt('density_water.dat')
x2 = data2[:,0]
y2 = data2[:,1]
coeff1 = np.polyfit(x2,y2,1)
coeff2 = np.polyfit(x2,y2,2)
p1 = np.poly1d(coeff1)
p2 = np.poly1d(coeff2)

print(p) # prints the polynomial expression
y1_fitted = p1(x2)
y2_fitted = p2(x2)
fig = plt.figure()
ax2 = fig.add_subplot(1, 1, 1)
ax2.plot(x2, y2, 'r')
ax2.plot(x2, y1_fitted, 'b',dashes=[1, 8])
ax2.plot(x2,y2_fitted, 'y')




```
